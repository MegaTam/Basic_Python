```python
import random
```

```python
class ecosystem:
    def __init__(self):
        self.lengthOfRiver()
        self.numberOfFish()
        self.numberOfBear()
        self.numberOfSteps()
        self.random()
        self.stimulate()

    def lengthOfRiver(self):
        while True:
            try:
                l = int(input('Please enter the length of river: '))
                if l > 0:
                    self.l = l
                    break
                else:
                    print('Please enter a positive integer')
            except:
                print('Please enter a positive integer')

    def numberOfFish(self):
        while True:
            try:
                f = int(input('Please enter the number of fish: '))
                if 0 <= f < self.l:
                    self.fish = f
                    break
                else:
                    print('Please enter a proper number: ')
            except:
                print('Please enter a proper number: ')

    def numberOfBear(self):
        while True:
            try:
                b = int(input('Please enter the number of bear: '))
                if 0 <= b <= (self.l - self.fish):
                   self.bear = b
                   break
                else:
                    print('Please enter a proper number: ')
            except:
                print('Please enter a proper number: ')

    def numberOfSteps(self):
        while True:
            try:
                s = int(input('Please enter the number of steps: '))
                if s > 0:
                    self.step = s
                    break
                else:
                    print('Please enter a positive integer: ')
            except:
                print('Please enter a positive integer: ')
            break

    def random(self):
        N = self.l - self.fish - self.bear
        self.initial = ['F']*self.fish + ['B']*self.bear + ['N']*N
        random.shuffle(self.initial)
        print('\nThe initial river is: ')
        for i in self.initial:
            print(i, end = ' ')
        print('\n')

    def change(self, p):
        while True:
            s = random.choice([-1, 0, 1])
            n = p + s
            if 0 <= n <len(self.initial):
                break
        if s == 0:
            return None
        if self.initial[n] == 'N':
            t = self.initial[p]
            self.initial[p] = self.initial[n]
            self.initial[n] = t
        else:
            animal = self.initial[p]
            if animal == self.initial[p]:
                if 'N' in self.initial:
                    pick = random.choice([k for k in range(len(self.initial)) if self.initial[k] == 'N'])
                    self.initial[pick] = animal
                else:
                    return True
            else:
                self.initial[p] = 'N'
                if animal == 'B':
                    self.initial[n] = 'B'
                    if s ==1 and (p+1) in self.animal:
                        self.animal.remove(p+1)

    def stimulate(self):
        for i in range(self.step):
            print('Step:', i+1)
            self.animal = [k for k in range(len(self.initial)) if self.initial[k] != 'N']
            for p in self.animal:
                self.change(p)
            for i in self.initial:
                print(i, end=' ')
            print('\n')
```

```python
run = ecosystem()
```
